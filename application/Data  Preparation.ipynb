## <span style="color:blue">Data Cleaning - df_main

```python
df_main = pd.read_stata("D:\Data Sampel Final 2022-20230608T163803Z-001\Data Sampel Final 2022\Kontekstual TB\df_main.dta")
```

```python
df_main['PNK03'] = pd.to_datetime(df_main['PNK03'], errors='coerce')
df_main['PNK04'] = pd.to_datetime(df_main['PNK04'], errors='coerce')
```

```python
# Mengisi nilai yang hilang dengan rata-rata untuk kolom numerik
df_main['PNK03'] = df_main['PNK03'].fillna(df_main['PNK03'].bfill())
df_main['PNK04'] = df_main['PNK04'].fillna(df_main['PNK04'].bfill())
df_main['FrekuensiKunjunganFKTP'] = df_main['FrekuensiKunjunganFKTP'].fillna(df_main['FrekuensiKunjunganFKTP'].bfill())

# Menghapus value duplikat di kolom sub data utama PSTV01
df_main = df_main.drop_duplicates(subset=['PSTV01'])


# Menampilkan hasil 
df_main.head()
```

## <span style="color:blue">Labeling Data

```python
df_main['PSTV03'] = pd.to_datetime(df_main['PSTV03'], errors='coerce')
df_main['Usia'] = (pd.to_datetime('today') - df_main['PSTV03']).dt.days // 365
```

```python
# 3. Frekuensi Kunjungan FKTP (jumlah kunjungan di FKTP)
bins = [0, 3, 10, float('inf')]
labels = ['Rendah', 'Sedang', 'Tinggi']
df_main['Frekuensi_Kunjungan_FKTP_Label'] = pd.cut(df_main['FrekuensiKunjunganFKTP'], bins=bins, labels=labels, right=False)
```

```python
# 4. Durasi Rawat FKRTL (menghitung durasi rawat di FKRTL)
bins_durasi = [0, 5, 14, float('inf')] 
labels_durasi = ['Pendek', 'Sedang', 'Panjang'] 
df_main['DurasiRawat'] = pd.cut(df_main['DurasiRawat'], bins=bins_durasi, labels=labels_durasi, right=False)
```

```python
# 5. Frekuensi Kunjungan FKRTL (jumlah kunjungan di FKRTL)
bins_fkr = [0, 5, 10, 20, float('inf')]
labels_fkr = ['Rendah', 'Sedang', 'Tinggi', 'Sangat Tinggi']  
df_main['Frekuensi_Kunjungan_FKRTL_Label'] = pd.cut(df_main['FrekuensiKunjunganFKRTL'], bins=bins_fkr, labels=labels_fkr, right=False)
```
